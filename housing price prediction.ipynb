## feature normalization function

```python
# Feature normalization(FOR DATAFRAME, NOT NUMPY ARRAY)
def featureNormalize(X, mu = 0, sigma = 0):
    if((type(mu)==int or type(mu)== int) and (mu==0 or sigma==0)):
        mu = np.mean(X)
        sigma = np.std(X)
        XNorm = (X - mu)/sigma
        return(XNorm, mu, sigma)
    else:
        XNorm = (X - mu)/sigma
        return(XNorm)
```

## linear regression regularized cost function

```python
# DATAFRAME Linear reg COST FUNCTION 
#seems all advanced optimizations flatten out thetas/x0, so necessary to RESHAPE theta inside cost fun
#takes biased X and return cost
def linearRegCostFun(theta, X, y, lam):
    #takes in theta, biased X, y and lam, return cost and gradients.
    #NOTE theta, if is feeded from optimization function, is flattened in shape(n, )
    #So IT IS NECESSARY TO RESHAPE THETA!!!
    [m,n] = X.shape
    theta = theta.reshape((n, 1))

    temp = np.dot(X, theta) - y
    J = 1/(2*m)*np.dot(temp.T, temp) + lam/(2*m) * np.sum(theta[1:]**2)

    grad = 1/m*np.dot(X.T,(np.dot(X, theta)-y)) + lam/m*theta
    grad[0] = 1/m*np.dot(X.T.iloc[0, :],(np.dot(X, theta)-y))
    return(float(J), grad.flatten())
```

```python
#linear regression theta can be initialized to 0
def trainLinearReg(X, y, lam):
    theta = np.zeros((X.shape[1], 1))
    options={'disp': False, 'gtol': 1e-6} #'gtol':1e-6, 'maxIters': 50,     for BFGS
    minimizeArgs = (X, y, lam)
    res = minimize(linearRegCostFun, theta, method='BFGS', jac=True, options=options, args=minimizeArgs) # callback = costTrack,
    return(res)

def costTrack(x):
    # CHANGE linearRegCostFun whenever use this function
    cost.append(linearRegCostFun(x, XPoly, trainy, lam)[0])
```

# learning curve

```python
def learningCurve(X,y,Xval,yval, lam):
    m = X.shape[0]
    errorTrain = np.zeros((m,1))
    errorVal = np.zeros((m,1))
    
    for i in range(1, m):
        print(str(i), end='\r')
        theta = trainLinearReg(X.iloc[:i, :], y.iloc[:i], lam).x
        errorTrain[i] = linearRegCostFun(theta, X.iloc[:i, :], y.iloc[:i], 0)[0]
        errorVal[i] = linearRegCostFun(theta, Xval, yval, 0)[0]
    return({'errorTrain': errorTrain, 'errorVal':errorVal})
```

# poly features

```python
def polyFeatures(X, p):
    #takes in unbiased X DataFrame, return p poly mapped DataFrame
    XPoly = X
    for i in range(2, p+1):
        temp = X**i
        temp.columns = X.columns + '^' + str(i)
        XPoly = pd.concat([XPoly, temp], axis = 1)
    return(XPoly)
```

```python
#1 lamvec creation
lamVec = np.zeros((24))
lamVec[0] = 0.00001
lamVec[1] = 0.00002
for i in range(2,len(lamVec)):
    lamVec[i] = lamVec[i-1]*2

#2. degree and create models
# XDict structure:
#XDict ->degree ->{XPoly, XPolyVal, XPolyTest, mu, sigma}
p = 20
XDict = {}

for i in range(1, p+1):
    #Xpoly configuration
    XPoly = polyFeatures(trainX, i)
    XPoly, mu, sigma = featureNormalize(XPoly)
    x0 = pd.DataFrame(np.ones((mTrain, 1)), columns=["x0"], index=XPoly.index)
    XPoly = pd.concat((x0, XPoly), axis = 1)


    #XpolyTest configuration
    XPolyTest = polyFeatures(testX, i)
    XPolyTest = featureNormalize(XPolyTest, mu, sigma)
    x0 = pd.DataFrame(np.ones((mTest, 1)), columns=["x0"], index=XPolyTest.index)
    XPolyTest = pd.concat((x0, XPolyTest), axis = 1)

    #XpolyVal configuration
    XPolyVal = polyFeatures(CVX, i)
    XPolyVal = featureNormalize(XPolyVal, mu, sigma)
    x0 = pd.DataFrame(np.ones((mCV, 1)), columns=["x0"], index=XPolyVal.index)
    XPolyVal = pd.concat((x0, XPolyVal), axis = 1)
    
    XDict[i]={'XPoly':XPoly, 'XPolyTest':XPolyTest, 'XPolyVal':XPolyVal, 'mu':mu, 'sigma':sigma}
    
# 3. lambda calculation and cost calculation CV and TEST
# 4. compute CV and train costwith all degrees, lambdas and their thetas.
# costDict structure
#costDict->degree->lambda->{theta, trainCost, CV cost, test cost}
costDict = {}
for i in range(1, p+1):
    costDict[i] = {}
    XPoly = XDict[i]['XPoly']
    XPolyTest = XDict[i]['XPolyTest']
    XPolyVal = XDict[i]['XPolyVal']
    train = np.zeros(len(lamVec))
    CV = np.zeros(len(lamVec))
    for j in range(len(lamVec)):
        costDict[i][lamVec[j]] = {}
        costDict[i][lamVec[j]]['theta'] = trainLinearReg(XPoly, trainy, lamVec[j]).x

        train[j] = linearRegCostFun(costDict[i][lamVec[j]]['theta'], XPoly, trainy, 0)[0]
        CV[j] = linearRegCostFun(costDict[i][lamVec[j]]['theta'], XPolyVal, CVy, 0)[0]
    costDict[i]['train'] = train
    costDict[i]['CV'] = CV
```
